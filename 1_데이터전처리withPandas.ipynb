## - DataFrame이란?
### DataFrame은 pandas에서 제공하는 **2차원 표 형식의 데이터 구조**로, Excel의 시트나 SQL 테이블과 유사합니다. DataFrame은 **행(row)**과 **열(column)**로 이루어져 있으며, 각 열은 서로 다른 데이터 타입(숫자, 문자열, 날짜 등)을 가질 수 있습니다.

## - Series란?
### Series: **1차원 배열**, **DataFrame의 각 열**에 해당하는 데이터
예: df['column_name']은 DataFrame의 특정 열을 나타내는 Series 객체입니다.

# Series 실습하기! => 번외) 추가공부

```python
import pandas as pd
```

```python
# 1. Series 생성하기_리스트를 통해! => pandas.Series()
data = [10,20,30,40,50]
series = pd.Series(data)
print(series)
```

## Why do we use index??

1. 데이터 조회
2. 데이터 정렬

```python
# 2. 인덱스 커스텀 가능! => pandas.Series(data , index = <== 여기!)
series_custom_index = pd.Series(data, index = ['a','b','c','d','e'])
print(series_custom_index)
```

```python
# 3. Series의 기본 연산 => Series는 벡터화 연산이 가능하다!
series_1 = pd.Series([1,2,3,4,5,])
series_2 = pd.Series([5,4,3,2,1,])
print(series_1 + series_2)
print(series_1 - series_2)
print(series_1 * series_2)
print(series_1 / series_2)
```

```python
# 4. Series에 접근하기. => 인덱스를 활용해서 특정값을 추출, 조작 가능하다!
series = pd.Series([1,2,3,4,5,6,7], index = ['a','b','c','d','e','f','g'])
print(series['a'])
print(series[series>5]) # =>>> 불린series를 불린마스크로 사용하여 데이터 필터링.
print(series>5)
```

## 불린 마스크란?

- 조건을 기준으로 **True 또는 False 값을 가지는 series를 생성**한다.

- 그 series를 원본 데이터에 인덱싱으로 적용하면, **True에 해당하는 값들만** 필터링할 수 있다.

```python
# 4_4. print(series[series>5]) 상세설명
print("불린 series")
print(series>5,end="\n\n") # => 이걸로 불린series를 생성.
print("불린 마스크를 통해 필터링된 값들")
print(series[series>5])
```

```python
# 5. Series의 메서드 사용 sum(), mean(), max()
s = pd.Series([1,2,3,4,5])
print(s.sum())
print(s.mean())
print(s.max())
print(s.min())
print(s.median()) # 중앙값
```

# DataFrame 이란? => 또 번외) 추가공부
---

## 1. DataFrame의 기본 개념 (Series를 곁들인)
- **Series**는 1차원 배열로, 인덱스와 값이 쌍을 이루는 구조입니다. 예를 들어, Series는 주로 한 변수에 대한 데이터를 저장하는 데 사용됩니다.

- **DataFrame**은 2차원 표 형식 데이터로, 여러 개의 Series가 컬럼으로 모여 있는 구조입니다. 즉, DataFrame은 여러 Series를 행과 열로 구성하여 표 형식으로 데이터를 다룰 수 있습니다.


---


## 2. Series에서 DataFrame으로의 확장

- Series와 DataFrame은 매우 밀접하게 연결되어 있어요. DataFrame의 각 열은 Series로 생각할 수 있습니다. 각 Series는 고유의 인덱스를 가지고 있고, DataFrame은 여러 개의 Series를 합친 형태입니다.

---


```python
import pandas as pd
# 데이터프레임 만들기 (그냥!!)
my_df = pd.DataFrame({
    'Age':[24,27,22],
    'Height':[5.5,5.8,5.6],
    'Weight':[55,56,57]
}, index=['Alice','Bob','Charlie'])
my_df
```

```python
# 데이터프레임 만들기 (인덱스가 있는 Series를 활용해서!)
series_Age = pd.Series([24,27,22],index=['Alice','Bob','Charlie'])
series_Height = pd.Series([5.5,5.8,5.6],index=['Alice','Bob','Charlie'])
series_Weight = pd.Series([55,56,57],index=['Alice','Bob','Charlie'])
my_df_with_Series = pd.DataFrame({
    'Age': series_Age,
    'Height' : series_Height,
    'Weight' : series_Weight
})
my_df_with_Series
```

## iloc와 loc에 대하여 알고 넘어가기!
- iloc: 정수 인덱스(데이터프레임의 **순서**!!)를 통한 접근
- loc: 내가 지은 이름(**데이터프레임에 써져있는 이름**!!)를 통한 접근 -> 이것도 숫자일 수 있다.

### iloc[<=== 행 순서 ===>,<=== 열 순서 ===>]

### loc[<=== 행 이름 ===>,<=== 열 이름 ===>]

### 만약 loc[<= 쉼표 없이 하나만 =>]  썼을 때는 '행'에 접근한다!!

---
## + 왜 df.iloc[2]에서 왜 대괄호 []를 사용하는지?
### => iloc는 함수가 아니라 속성(attribute)로, 인덱서(indexer)라고 불리는 특별한 객체이다.
### 따라서 "**파이썬 리스트나 배열에서 데이터를 인덱싱하거나 슬라이싱할 때 대괄호 []를 사용하는 것처럼**"

### iloc는 DataFrame이나 Series에서 []를 통해 정수 기반 인덱싱을 할 수 있도록 돕는 인덱서다.

## 필터링 1. 불린 시리즈 만들기

불린 시리즈는 True 또는 False 값으로 이루어진 시리즈입니다. 이는 조건에 대한 결과를 담고 있습니다. 예를 들어, 특정 열에 대해 조건을 평가하고 그 결과를 불린 값으로 나타낸 것이 바로 불린 시리즈입니다.

```python
friend_df['age']>30
```

## 필터링 2. 불린 마스크

불린 마스크는 **불린 시리즈를 DataFrame에 인덱싱**할 때 사용되는 표현입니다. 불린 시리즈를 통해 True 값이 있는 인덱스만 선택하는 역할을 합니다.

```python
friend_df[friend_df['age']>30]
# friend_df[[False, False, False, True, True, False]] 와 동일!
```

```python
# Q. job이 'intern'인 사람을 가져와라.
friend_df[friend_df['job']=='intern']
```

# 조건 여러개를 처리할 때

## 파이썬에서의 and, &&, &
- and: 파이썬에서 "**단일 조건**"을 결합할 때 사용합니다.
- &&: &&는 파이썬에서 사용할 수 없습니다. C, C++, Java, JavaScript 등에서는 &&를 사용하지만, 파이썬에서는 대신 and를 사용합니다.
- &: &는 비트 단위 연산자입니다. &는 각 조건을 "**element-wise로 비교**"하여, 각 조건이 True인 곳에서 True를 반환하는 역할을 합니다.




```python
# Q. 30대 이상, 40대 이하인 사람들을 가져와라.
friend_df[(friend_df['age']>=30) & (friend_df['age']<=40)]
```

```python
# Q. 30대 미만 혹은 40대 초과인 사람들을 가져와라.
friend_df[(friend_df['age']<30) | (friend_df['age']>40)]
```

# in을 통한 포함 조건 걸기 (람다 함수를 통한)

- in 왼쪽은 단일값이어야 한다!!!

```python
# ex
5 in [1,3,4,5,6,7,8,10]
```

```python
# 오류생김: The truth value of a Series is ambiguous.
## df[df['job'] in ['student','manager']]

# 5는 단일 값이지만, df['job']는 시리즈 값이므로.
# in을 사용하려면 in 왼쪽은 단일값이어야 한다!!!
```

```python
### 정답!!! 뒤에 설명!
friend_df[friend_df['job'].apply(lambda x : x in ['student','manager'])]
```

# apply()
- apply()는 pandas의 메서드로, Series나 DataFrame의 각 원소에 대해 사용자 정의 함수(혹은 람다 함수)를 적용할 수 있게 해줍니다.



# lambda
- lambda는 익명 함수를 정의할 때 사용하는 파이썬의 기능입니다.
- lambda x: x in ['student', 'manager']는 x가 'student' 또는 'manager'인지 확인하는 함수입니다.
