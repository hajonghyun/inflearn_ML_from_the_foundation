## header 매개변수의 사용법
- header=0 (기본값): 첫 번째 행을 헤더로 사용.
- header=None: 헤더가 없다고 간주하고, 첫 번째 행을 데이터로 읽고, 자동으로 숫자형 열 이름(0, 1, 2, ...)을 할당.
- header=n: n번째 행을 헤더로 사용 (0부터 시작하는 인덱스).
- header=[n, m]: 여러 행을 헤더로 사용하여 멀티 인덱스를 만들 수 있음.

## sep 매개변수의 사용법

## groupby 함수의 기본 작동 원리
1. 데이터를 지정된 기준에 따라 나누고,
2. 그룹화된 각 부분에 대해 연산(합계, 평균, 최대값, 최소값)을 수행합니다.

````python
# groupby 함수를 통해 성별에 따른 'whole weight'의 합을 확인해보자.
# 방법1: series에 groupby 적용하기
grouped = abalone_df['whole_weight'].groupby(abalone_df['sex'])
grouped.sum()
# 주의점!!! sum()값이 크다고 해서 male의 weight가 대체적으로 크다고 생각해서는 안된다.
# 단순히 male 개체 수가 많을 수도 있기 때문이다. 평균을 통해 알아보자.
````

````python
# 방법2: 데이터프레임에 groupby 적용하기
abalone_df.groupby('sex')['whole_weight'].sum()
````

````python
# groupby 함수를 통해 성별에 따른 'whole weight'의 평균을 확인해보자.
grouped.mean()
# 이처럼 Female의 weight가 대체적으로 무거움을 알 수 있다.
# 하지만 이상치가 포함되어있을 수 있기 때문에 시각화를 통해 정확히 알아봐야한다.
````

````python
# groupby 후(각 성별)의 데이터 개수도 알 수 있다.
grouped.size()
````

````python
# 이번엔 그룹변수가 하나('whole weight')가 아닌, 전체 연속형 변수에 대한 집계를 해보자.
abalone_df.groupby(abalone_df['sex']).mean()
````

````python
# 위의 코드는
abalone_df.groupby('sex').mean()
# 와 같이 단순화 할 수 있다.
````

---
## 단순화할 때 들었던 의문사항 해결!!
위의
```
grouped = abalone_df['whole_weight'].groupby(abalone_df['sex'])
```
는 다음과 같이 단순화할 수 "없다".
```
grouped = abalone_df['whole_weight'].groupby('sex')
```
왜냐하면 groupby 함수가 Series 객체에서 호출되었기 때문입니다.
- abalone_df['whole_weight']는 Series 객체입니다. 이때, groupby는 Series에 대해 그룹화 기준(여기서는 'sex') 을 적용하려고 합니다.
-abalone_df['whole_weight']는 이미 하나의 컬럼(Series)이기 때문에, 'sex'라는 컬럼을 그룹화 기준으로 제대로 사용할 수 없습니다.
---

## 하지만!
```
abalone_df.groupby(abalone_df['sex']).mean()
```
- 여기서는 groupby가 데이터프레임 전체에 적용되는 방식입니다.

따라서
```
abalone_df.groupby('sex').mean()
```
이렇게 단순화가 가능하다.
---
---


## 정리
- Series에서의 groupby:
  - abalone_df['whole_weight'].groupby(abalone_df['sex'])는 Series에서만 사용됩니다. 즉, 특정 컬럼에 대해서만 그룹화하고자 할 때 groupby를 사용합니다.

  - 여기서 sex는 그룹화의 기준이 아니라, 해당 Series의 인덱스 역할을 하게 되어 오류가 발생할 수 있습니다.

- DataFrame에서의 groupby:
  - abalone_df.groupby('sex')는 데이터프레임에서 행 단위로 그룹화할 때 사용됩니다. 이때는 sex 컬럼을 기준으로 전체 데이터를 그룹화한 뒤, 그룹별로 연산(평균, 합계 등)을 수행할 수 있습니다.

````python
# np.where()를 사용해 새로운 조건('length'가 중간값보다 크면 'length_long' 작으면 'length_short')
# 에 맞는 변수('length bool')를 추가하자.
import numpy as np

abalone_df['length_bool'] = np.where(abalone_df['length']>abalone_df['length'].median(),
                                     'length_long',  #중간값보다 크면
                                     'length_short') #중간값보다 작으면
abalone_df['length_bool']
````

````python
# 그룹변수를 2개 이상('sex','length_bool') 선택해서 통계처리(mean)를 해보자
abalone_df.groupby(['sex','length_bool']).mean()
````

````python
# 그룹변수를 2개 이상('sex','length_bool') 선택하고
# 'whole_weight'에 대한 mean값 구하기
abalone_df.groupby(['sex','length_bool'])['whole_weight'].mean()
````

### fillna() 함수 사용법
fillna() 함수의 가장 기본적인 사용법은 결측치를 특정 값으로 채우는 것입니다.
1. 특정 값으로 결측치 채우기

    ``` df.fillna(0)```

    모든 NaN 값을 0으로 대체합니다.

2. 각 열에 대해 다른 값으로 결측치 채우기

    ``` df.fillna({'col1': 0, 'col2': 5})```
    
    col1 열의 NaN 값은 0으로, col2 열의 NaN 값은 5로 대체합니다.

3. 앞의 값으로 결측치 채우기 (ffill)

    ``` df.fillna(method='ffill')```
    
    각각의 NaN 값을 앞의 값으로 채웁니다.

4. 뒤의 값으로 결측치 채우기 (bfill)

    ``` df.fillna(method='bfill')```
    
    각각의 NaN 값을 뒤의 값으로 채웁니다.

5. 특정 열에 대해 결측치 채우기
    
    ```df['col1'] = df['col1'].fillna(0)```
    
    특정 열에 대해서만 NaN 값을 채웁니다.

6. 행 또는 열 단위로 결측치 채우기 (axis)

    ```df.fillna(method='ffill', axis=1)```

    각 행을 따라 NaN 값을 채웁니다.

7. inplace로 원본 수정

    ```df.fillna(0, inplace=True)```

    이렇게 하면 df 객체 자체가 수정됩니다. inplace=True를 설정하면 새로운 객체를 반환하지 않고, 원본 객체가 변경됩니다.

8. limit으로 채우는 개수 제한

    ```df.fillna(0, limit=1)```
    
    결측치를 채우는 개수를 1개로 제한합니다. 이 경우 첫 번째 NaN 값만 채워지고, 나머지 값은 그대로 두어집니다.

````python
import math
def avg_ceil(x,y,z):
    return math.ceil((x+y+z)/3)

abalone_df[['diameter','height','whole_weight']]
````

````python
abalone_df[['diameter','height','whole_weight']].apply(
    lambda x : avg_ceil(x[0],x[1],x[2]) , axis=1
)
````

````python
# 문제
# 1. 사용자 정의 함수 사용
# 2. ['diameter','height','whole_weight'] 변수 사용
# 3. 세 변수의 합이 1이 넘으면 True, 아니면 False 출력 후 answer 변수에 저장
# 4. abalone_df에 answer 열을 추가하고 입력
````

````python
from ast import Lambda
def over_one(x,y,z):
    if x+y+z>1:
        return True
    else:
        return False

abalone_df[['diameter','height','whole_weight']].apply(
    lambda x: over_one(x[0],x[1],x[2]), axis=1
)
````

````python
my_bool_df = abalone_df[['diameter','height','whole_weight']].apply(
                lambda x: over_one(x[0],x[1],x[2]), axis=1
            )
abalone_df['answer'] = my_bool_df
abalone_df
````
