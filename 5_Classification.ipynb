## 1️⃣ **이론: 로지스틱 회귀는 확률 예측 모델이다**

로지스틱 회귀는 입력 벡터 `x`에 대해 **특정 클래스(여기서는 virginica)에 속할 확률**을 예측합니다. 수식은 다음과 같습니다:

```
p̂(x) = σ(wᵗx + b) = 1 / (1 + exp(-(wᵗx + b)))
```

여기서:
- `w`: 모델이 학습한 가중치
- `b`: 편향
- `σ`: 시그모이드 함수
- `p̂(x)`: 입력 `x`에 대해 virginica일 확률

즉, 입력값 `x`를 선형 변환하여 시그모이드 함수에 넣으면 0~1 사이의 확률값이 나옵니다.

---

## 2️⃣ `y_proba = log_reg.predict_proba(x_new)`의 의미

- `x_new`는 0부터 3까지 균등한 1000개의 값을 가진 1D 입력 (예: 꽃잎 너비)
- `log_reg.predict_proba(x_new)`는 각 입력값에 대해 `[클래스 0일 확률, 클래스 1일 확률]`을 반환합니다.

즉, `y_proba[:, 1]`은 각 `x`에 대해 "이 샘플이 virginica일 확률"을 시그모이드 함수에 의해 계산한 결과입니다.

이 확률은 다음과 같은 수식을 통해 모델 내부에서 계산됩니다:

```
y_proba[i, 1] = 1 / (1 + exp(-(w * x_new[i] + b)))
```

이것이 시그모이드 함수 그 자체이며, 모델이 판단한 클래스 1 (virginica)일 가능성을 의미합니다.

---

## 3️⃣ 시각화된 곡선은 무엇을 보여주는가?

```python
plt.plot(x_new, y_proba[:, 1], 'r-', label='virginica')
```

이 코드는 x 값(꽃잎 너비)가 증가할수록 virginica일 확률이 어떻게 바뀌는지를 곡선으로 보여줍니다:

- 확률이 **0.5인 지점**이 **결정 경계(decision boundary)**입니다.
  - 예: 꽃잎 너비가 1.75cm 이상이면 virginica일 가능성이 50%를 넘음
- 확률이 1에 가까워질수록 모델은 "이건 확실히 virginica"라고 판단

즉, 이 곡선은 로지스틱 회귀 모델의 시그모이드 결정 경계를 시각화한 것입니다.

---

## ✅ 최종 요약

| 구성 요소 | 설명 |
|-----------|------|
| `x_new` | 꽃잎 너비가 0~3 사이일 때의 입력값들 |
| `predict_proba(x_new)` | 각각의 x값에 대해 virginica일 **확률**을 시그모이드로 계산 |
| `y_proba[:,1]` | 그 확률값만 추출 (virginica일 확률) |

이처럼 `predict_proba`에서 나온 값은 시그모이드 함수의 출력값이며, 곡선은 로지스틱 회귀가 어떻게 입력값을 기반으로 확률적 판단을 내리는지를 보여줍니다.

___
# Softmax 회귀 (다항 logistic 회귀)

| 항목       | 로지스틱 회귀 (이항 분류)                                    | 소프트맥스 회귀 (다중 클래스 분류)                                                      |
| -------- | -------------------------------------------------- | ------------------------------------------------------------------------- |
| 분류 대상    | 두 개의 클래스 (예: 0 vs 1)                               | 세 개 이상 클래스 (예: 0, 1, ..., K)                                              |
| Score 계산 | $s(\mathbf{x}) = \boldsymbol{\theta}^T \mathbf{x}$ | $s_k(\mathbf{x}) = (\boldsymbol{\theta}^{(k)})^T \mathbf{x}$ for each $k$ |
| 확률 계산    | $\hat{p} = \sigma(s) = \frac{1}{1 + \exp(-s)}$     | $\hat{p}_k = \frac{\exp(s_k)}{\sum_{j=1}^{K} \exp(s_j)}$                  |
| 출력값      | 클래스 1일 확률 (0\~1)                                   | 각 클래스 $k$에 대한 확률 (총합 1)                                                   |
| 최종 예측    | $\hat{y} = 1$ if $\hat{p} > 0.5$                   | $\hat{y} = \arg\max_k \hat{p}_k$                                          |
| 손실함수     | Binary Cross-Entropy                               | Categorical Cross-Entropy (Softmax Loss)                                  |

### 2. **출력값 해석**

- 로지스틱 회귀는 **“한 쪽(class 1)일 확률”**만 예측
- 소프트맥스 회귀는 **“각 클래스가 정답일 확률”**을 동시에 출력

---

### 3. **결정 기준**

- 로지스틱 회귀는 단일 임계값(0.5 기준)으로 0/1 결정
- 소프트맥스 회귀는 가장 높은 확률을 가진 클래스를 선택

---

## 🧠 기억하면 좋은 점

- 소프트맥스 회귀는 로지스틱 회귀의 일반화된 형태로, **K개의 클래스 각각에 대해 score를 계산한 후 softmax로 정규화**하는 방식
- 로지스틱 회귀는 사실상 softmax 회귀의 \( K=2 \)일 때의 특수한 경우로 볼 수 있음

````python
# 파이프라인 안에 만들것들
# 1. 변수변환: polynomial 3차 다항식
# 2. scaler: standardscaler
# 3. 모델: linearSVC

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.svm import LinearSVC

poly_std_svm = Pipeline([
    ("polynomial",PolynomialFeatures(degree=3)),
    ("std",StandardScaler()),
    ("svm",LinearSVC(C=10,loss='hinge'))
])

# 모델학습
poly_std_svm.fit(x,y)

#예측
new_moon = [[2.0,1.0]]
p = poly_std_svm.predict(new_moon)
print(p)
````

````python
import numpy as np
# 결정 경계 시각화 (meshgrid 없이)
x_n = np.linspace(-1.5, 2.5, 300)
y_n = np.linspace(-1.0, 1.5, 300)
points = np.array([[a, b] for a in x_n for b in y_n])  # 모든 조합을 직접 생성
zz = poly_std_svm.predict(points).reshape(len(x_n), len(y_n))
print(zz)
plt.contourf(x_n, y_n, zz.T, alpha=0.3)  # .T로 x/y 방향 정렬
plt.scatter(x[:, 0], x[:, 1], c=y, edgecolor='k')
plt.show()
````

# 커널

### 커널 트릭(Kernel Trick)
- 기본 원리: 실제로 데이터를 고차원 공간으로 변환하지 않고, 커널 함수를 통해 고차원 내적을 직접 계산할 수 있습니다. 예를 들어, 선형 분리 불가능한 데이터를 고차원 공간으로 올리면 선형 분리가 가능해지는 경우, 명시적 특징 매핑 대신 커널 함수를 사용하여 연산 효율성을 높입니다.

- 주요 응용: 서포트 벡터 머신(SVM), 커널 주성분 분석(Kernel PCA), 커널 리지 회귀 등 다양한 알고리즘에서 활용됩니다.

````python
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

poly_kernel_std_svm = Pipeline([
    ('std',StandardScaler()),
    ('poly_kernel_svm',SVC(kernel='poly',degree=3,coef0=1,C=5))
])
poly_kernel_std_svm.fit(x,y)
````

# 가우시안 RBF 커널

````python
rbf_kernel_std_svm = Pipeline([
    ('std',StandardScaler()),
    ('rbf_kernel_svm',SVC(kernel='rbf',gamma=3,C=0.001))
])
rbf_kernel_std_svm.fit(x,y)
````
